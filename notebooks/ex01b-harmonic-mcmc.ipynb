```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from time import time
```

```python
def phi_pot(xy):
    return (xy[0]**2 + xy[1]**2)/2
```

```python
def MCMC_random_walk_in_harmonic_potential(steps, width):
    xy = np.zeros((steps,2))
    xy[0] = width*np.random.randn(2)
    dxy = width*np.random.randn(steps,2)
    for i in range(steps-1):
        xy_new = xy[i] + width*dxy[i]
        energy_ratio = np.exp(-phi_pot(xy_new))/np.exp(-phi_pot(xy[i]))
        acceptance_prob = min(1,energy_ratio)
        if np.random.rand(1) < acceptance_prob:
            xy[i+1] = xy_new
    return xy
```

```python
start = time()
xy = MCMC_random_walk_in_harmonic_potential(100000,0.1)
print(time()-start)
plt.plot(xy[:,0],xy[:,1],'--*')
plt.ylabel('some numbers')
plt.show()
```
